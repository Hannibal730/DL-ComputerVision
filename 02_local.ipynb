## numpy 배열 만들기
-1 부터 1까지 step(=0.1) 단위로 숫자 생성  
`np.arange`함수(array range)는 파이썬의 `range`와 비슷한 사용법이다.  


````python
import numpy as np
x = np.linspace(-1, 1, num=21)
x
````

````python
y = x**2
y
````

````python
print(x.shape, y.shape)
````

그래프 그려보기

````python
from matplotlib import pyplot as plt
plt.plot(x, y, '-')
plt.show()
````

## 텐서와 다차원 배열의 차이

텐서는 주로 정형화된 수치 배열에 사용한다.
텐서는 병렬 연산자를 적용할 수 있다.

````python
x = np.array([
    [ 1.0, 1.1 ],
    [ 2.2, 1.0 ]
])
y = [
    [ 1.0, 1.1 ],
    [ 1.0, 3, 2.0 ],

]
print(x.shape)
x = x*2
print(x)
x = x**2 + 1
print(x)

# 배열에 2를 곱하는 것은 다른 의미를 가짐.
y = y*2  ## same as y = y + y
print(y)
# 다음 연산은 오류가 남.
# y = y**2
# y = y + 1

print(x[0,1])
print(x[0][1])
# y[1,2] # 배열에는 tupule 형식을 사용할 수 없다.
````

# 확장하기

과일의 크기와 무게를 모두 2배로 하기

````python
import numpy as np
x = np.array([[1.0, 1.1], [2.0, 2.1], [0.9, 1.1]])
x_big = x*2
print(x_big)
````

무게만 두 배로 하기

````python
x_size = x[:, 1:2] # 두번째 열(column)만 추출
x_size *= 2
x_big2 = x
x_big2[:, 1:2] = x_size # 두번째 열 데이터 교체
print(x_big2)
````
